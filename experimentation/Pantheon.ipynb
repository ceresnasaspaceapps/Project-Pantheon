# **Map Generator**

```python
import folium
from folium.plugins import HeatMap
import requests
from collections import defaultdict

api_key = "cfc9efc1d0f159518f7c843fa92a0635"

def get_air_quality(lat, lon):
    url = f"http://api.openweathermap.org/data/2.5/air_pollution?lat={lat}&lon={lon}&appid={api_key}"
    response = requests.get(url)
    return response.json()['list'][0]['components']

def get_rain(lat, lon):
    url = f"http://api.openweathermap.org/data/2.5/weather?lat={lat}&lon={lon}&appid={api_key}"
    response = requests.get(url)
    return response.json().get('rain', {}).get('1h', 0)

def get_current_weather(lat, lon):
    url = f"http://api.openweathermap.org/data/2.5/weather?lat={lat}&lon={lon}&appid={api_key}"
    response = requests.get(url)
    data = response.json()
    return {
        'description': data['weather'][0]['description'],
        'temp': data['main']['temp'] - 273.15,
        'humidity': data['main']['humidity'],
        'clouds': data['clouds']['all']
    }

def icon_size(impact):
    base_size = 20
    scale_factor = 4
    return (int(base_size + scale_factor * impact), int(base_size + scale_factor * impact))

def student_contributions(locations):
    student_impact = defaultdict(float)
    for location in locations:
        student = location['name'].split(",")[0].strip()  # Use the student's name
        student_impact[student] += location['impact']
    return sorted(student_impact.items(), key=lambda x: x[1], reverse=True)

my_map = folium.Map(location=[36.7250, -4.4195], zoom_start=14)
icon_url = "https://cdn-icons-png.flaticon.com/512/4778/4778362.png"

for location in locations:
    air_quality = get_air_quality(location['lat'], location['lon'])
    rain = get_rain(location['lat'], location['lon'])
    weather = get_current_weather(location['lat'], location['lon'])
    size = icon_size(location['impact'])

    icon = folium.CustomIcon(
        icon_image=icon_url,
        icon_size=size,
        icon_anchor=(size[0]//2, size[1])
    )

    popup_html = f"""
    <div style="font-size: 12px; width: 150px;">
    <strong>{location['name']}</strong><br>
    Air Quality AQI PM2.5: {air_quality['pm2_5']} Âµg/mÂ³<br>
    Rain: {rain} mm/h<br>
    Weather: {weather['description']}<br>
    Temp: {weather['temp']:.2f}Â°C<br>
    Humidity: {weather['humidity']}%<br>
    Cloudiness: {weather['clouds']}%<br>
    Trees planted: {location['impact']}
    </div>
    """

    folium.Marker(
        location=[location['lat'], location['lon']],
        popup=folium.Popup(popup_html, max_width=265),
        icon=icon
    ).add_to(my_map)

sorted_students = student_contributions(locations)

ranking_html = "<h4 style='font-family:Arial, sans-serif; color:#2C3E50; text-align:center;'>Ranking of students by impact</h4><ul style='list-style-type:none; padding:0;'>"
for i, (student, impact) in enumerate(sorted_students, start=1):
    ranking_html += f"<li style='padding:10px 0; border-bottom:1px solid #ccc;'><strong>{i}. {student}</strong>: <span style='color:#27AE60;'>{impact} pt</span></li>"
ranking_html += "</ul>"

sidebar_html = f"""
<div id="mySidebar" style="position: fixed; top: 0; right: -300px; width: 300px; height: 100%; padding: 20px; background-color: #f9f9f9; border-left: 3px solid #2980B9; box-shadow: -2px 0 10px rgba(0,0,0,0.1); z-index: 1000; transition: 0.5s; overflow-y: auto; border-radius: 10px 0 0 10px;">
    <a href="javascript:void(0)" style="text-decoration:none; font-size:30px; position:absolute;top:10px;left:15px;color:#E74C3C;" onclick="closeSidebar()">Ã</a>
    {ranking_html}
</div>
<div id="openButton" style="position:fixed;top:10px;right:10px;z-index:1001;">
    <button onclick="openSidebar()" style="padding:10px 15px;background-color:#2980B9;border:none;border-radius:5px;color:white;font-weight:bold;cursor:pointer;box-shadow:0 4px 6px rgba(0,0,0,0.1);">Show Ranking</button>
</div>
<script>
function openSidebar() {{
    document.getElementById('mySidebar').style.right = '0';
    document.getElementById('openButton').style.display = 'none';
}}
function closeSidebar() {{
    document.getElementById('mySidebar').style.right = '-300px';
    document.getElementById('openButton').style.display = 'block';
}}
</script>
"""

my_map.get_root().html.add_child(folium.Element(sidebar_html))
my_map.save("custom_icons_map.html")
my_map
```

#**Ejemplo visualizaciÃ³n de datos**

```python
import pandas as pd
import numpy as np

# NÃºmero de personas y registros por persona
n_personas = 20
n_registros_por_persona = 50

# Listado de nombres ficticios para las personas
nombres = [
    'InÃ©s SÃ¡nchez', 'JosÃ© MartÃ­nez', 'Carmen RodrÃ­guez', 'Antonio SÃ¡nchez', 'LucÃ­a FernÃ¡ndez',
    'MarÃ­a GÃ³mez', 'Francisco GarcÃ­a', 'Laura JimÃ©nez', 'Javier LÃ³pez', 'Raquel Torres',
    'David Moreno', 'Beatriz Ruiz', 'Daniel PÃ©rez', 'Marta Romero', 'Juan Torres',
    'Alicia Medina', 'AndrÃ©s Molina', 'Carlos Navarro', 'Pilar SuÃ¡rez', 'SofÃ­a GonzÃ¡lez'
]

# Crear un DataFrame vacÃ­o
data = []

# Fecha de inicio
start_date = pd.to_datetime("2023-01-01")

# Generar registros para cada persona
for nombre in nombres:
    # Valores iniciales para suavizar cambios entre registros
    temp_base = 10  # Base inicial mÃ¡s baja para temperatura
    humedad_base = 80  # Base inicial mÃ¡s alta para humedad
    aqi_base = 1.5  # Base inicial para AQI PM2.5

    # Variaciones estacionales
    for i in range(n_registros_por_persona):
        current_date = start_date + pd.DateOffset(weeks=i)
        month = current_date.month

        # Simular variaciÃ³n estacional de la temperatura
        if month in [12, 1, 2]:  # Invierno
            seasonal_temp_variation = np.random.uniform(-5, 5)
        elif month in [3, 4, 5]:  # Primavera
            seasonal_temp_variation = np.random.uniform(0, 10)
        elif month in [6, 7, 8]:  # Verano
            seasonal_temp_variation = np.random.uniform(15, 25)
        elif month in [9, 10, 11]:  # OtoÃ±o
            seasonal_temp_variation = np.random.uniform(5, 15)

        temperatura = temp_base + seasonal_temp_variation
        humedad = np.clip(humedad_base - 0.3 * seasonal_temp_variation, 20, 100)  # Ajuste inverso con la temperatura

        # AQI PM2.5 varÃ­a con la temperatura y la lluvia potencial
        lluvia = max(0, np.random.normal(2, 1))
        aqi_variation = np.random.uniform(-0.1, 0.1) + 0.05 * lluvia  # Aumenta ligeramente con mÃ¡s lluvia
        aqi_pm2_5 = np.clip(aqi_base + aqi_variation - 0.01 * temperatura, 0.5, 2.5)  # Menor en temperaturas altas

        registro = {
            'Nombre': nombre,
            'Fecha': current_date,
            'AQI_PM2.5 (Âµg/mÂ³)': aqi_pm2_5,
            'Lluvia (mm/h)': lluvia,
            'Temperatura (Â°C)': temperatura,
            'Humedad (%)': humedad,
            'Nubosidad (%)': np.random.uniform(0, 100),
            'Riego Semanal (puntos)': np.random.randint(5, 11),
            'PuntuaciÃ³n': np.random.randint(0, 11)  # CorrecciÃ³n aquÃ­ para incluir 0 en el rango
        }
        data.append(registro)

# Crear DataFrame a partir de los registros generados
df_grande = pd.DataFrame(data)

# Mostrar los primeros registros
print(df_grande.head(20))
```

### **We maintain a historical record of your plant's data, allowing us to track the weather conditions it has experienced, the number of times you've watered it, and the score associated with its care. This enables us to not only access real-time data but also analyze historical trends for a deeper understanding of your plant's health and care patterns.**

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from scipy.ndimage import gaussian_filter1d

# Set seaborn style
sns.set(style="whitegrid")

# Filter data for a specific person (e.g., 'InÃ©s SÃ¡nchez')
person = 'InÃ©s SÃ¡nchez'
df_person = df_grande[df_grande['Nombre'] == person]

# Create figure and axes
fig, axs = plt.subplots(3, 2, figsize=(14, 14), dpi=80)
fig.suptitle(f'Evolution of Conditions for {person}', fontsize=16, fontweight='bold')

# Format dates for better visualization
df_person['Fecha'] = pd.to_datetime(df_person['Fecha'])

# Smooth lines for better aesthetic visualization
def smooth_line(series):
    return gaussian_filter1d(series, sigma=2)

# AQI PM2.5
sns.lineplot(x='Fecha', y=smooth_line(df_person['AQI_PM2.5 (Âµg/mÂ³)']), data=df_person, ax=axs[0, 0], color='green', marker='o')
axs[0, 0].set_title('Air Quality (AQI PM2.5)', fontsize=14)
axs[0, 0].set_ylabel('Âµg/mÂ³')

# Rain
sns.lineplot(x='Fecha', y=smooth_line(df_person['Lluvia (mm/h)']), data=df_person, ax=axs[0, 1], color='blue', marker='o')
axs[0, 1].set_title('Rain', fontsize=14)
axs[0, 1].set_ylabel('mm/h')

# Temperature
sns.lineplot(x='Fecha', y=smooth_line(df_person['Temperatura (Â°C)']), data=df_person, ax=axs[1, 0], color='red', marker='o')
axs[1, 0].set_title('Temperature', fontsize=14)
axs[1, 0].set_ylabel('Â°C')

# Humidity
sns.lineplot(x='Fecha', y=smooth_line(df_person['Humedad (%)']), data=df_person, ax=axs[1, 1], color='purple', marker='o')
axs[1, 1].set_title('Humidity', fontsize=14)
axs[1, 1].set_ylabel('%')

# Weekly Watering
sns.lineplot(x='Fecha', y=smooth_line(df_person['Riego Semanal (puntos)']), data=df_person, ax=axs[2, 0], color='orange', marker='o')
axs[2, 0].set_title('Weekly Watering', fontsize=14)
axs[2, 0].set_ylabel('Points')

# Score
sns.lineplot(x='Fecha', y=smooth_line(df_person['PuntuaciÃ³n']), data=df_person, ax=axs[2, 1], color='brown', marker='x')
axs[2, 1].set_title('Score', fontsize=14)
axs[2, 1].set_ylabel('Points')
axs[2, 1].set_ylim(0, 10)  # Set Y-axis range from 0 to 10

# Adjust layout and display
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
